```python

import pandas as pd
import seaborn as sns
data = pd.read_csv('House Price India.csv')
print(data.head())
```

**Univariate** **Analysis**

```python
sns.histplot(data['number of bathrooms'])
```

**Bivariate** **Analysis**

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
sns.barplot(x=data['living area'], y=data['number of bathrooms'])
```

**Multivariate Analysis**

```python
df = data
dat=df[['number of bathrooms', 'number of bedrooms','living area']]
sns.heatmap(dat.corr(), annot=True)
```
